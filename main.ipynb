```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2.5 * X + np.random.randn(100, 1) * 5
plt.figure(figsize=(8, 6))
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Sample Data for Linear Regression (Synthetic Data)')
plt.show()
from sklearn.linear_model import LinearRegression   

```

```python
import pandas as pd

data = pd.DataFrame({'X': X.flatten(), 'y': y.flatten()})
data.to_csv('linear_regression_data.csv', index=True)
print("Data saved to 'linear_regression_data.csv'") 

print("Total number of rows in the dataset:", data.head(10))
print("Total number of rows in the dataset:", data.tail(10))
```
